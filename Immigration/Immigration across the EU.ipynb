# Immigration across EU from 2010 to 2019 (General Overview)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

* Importing the data as a DataFrame from the excel sheet

```python
#Importing the dataset from the excel file:
#Total Employment in STEM, in thousands: 
df=pd.read_excel("C:\\Users\\ibsi\\Desktop\\STEM-HTEC-Emplyoment-in-the-EU\\Datasets\\Immigration.xls", sheet_name="Total")
df.head(5)
```

```python
# Boxplot for the years 2011 to 2018:
boxplot = df.boxplot(column=["2012", "2013", "2014", "2015", "2016", "2017", "2018"])
plt.xlabel('Years', fontsize=10)
plt.ylabel('Immigrants', fontsize=10)
plt.savefig("C:\\Users\\ibsi\\Desktop\\Figure_12", transparent=True, bbox_inches="tight")
```

```python
#Calculating the cumulative (total) immigration per country:
Total=df.sum(axis=1)
df["Cumulative Immigration"]=Total
df.head(5)
```

```python
# Sorting the EU countries accodring to the cumulative immigration rate:
ax=df.sort_values('Cumulative Immigration', ascending=True).plot.barh(x="Country", y="Cumulative Immigration", figsize=(30,40))
plt.xlabel('Cumulative Immigration (in thousands)', fontsize=40)
plt.ylabel('Country', fontsize=40)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.legend(loc=4, prop={'size': 40})

for i in ax.patches:
    
    ax.text(i.get_width()+.1, i.get_y()+.31, \
            str(int((i.get_width()))), fontsize=30, color='dimgrey')
plt.savefig("C:\\Users\\ibsi\\Desktop\\Figure_14", transparent=True, bbox_inches="tight")    
```

```python
# Histogram showing the cumulative immigration:
plt.hist(df["Cumulative Immigration"], color = 'skyblue', edgecolor = 'black', lw=0.5)
plt.axvline(df["Cumulative Immigration"].mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(df["Cumulative Immigration"].median(), color='k', linestyle='solid', linewidth=1)
plt.xlabel('Cumulative Immigration (in thousands)', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.savefig("C:\\Users\\ibsi\\Desktop\\Figure_13", transparent=True, bbox_inches="tight")
```
