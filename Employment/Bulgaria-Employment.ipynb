# Employment in the HTEC/STEM sector (Bulgaria)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Total Employment in STEM (in thousands): 
df=pd.read_excel("C:\\Users\\ibsi\\Desktop\\STEM-HTEC-Emplyoment-in-the-EU\\Datasets\\Employment in HTEC.xlsx", sheet_name="HTEC(thousands)")
df=df.drop(df.index[0:4]).reset_index(drop = True)
df.head(5)
```

```python
#Employment with less than a Secondary education:
df1=pd.read_excel("C:\\Users\\ibsi\\Desktop\\STEM-HTEC-Emplyoment-in-the-EU\\Datasets\\Employment in HTEC.xlsx", sheet_name="<Secondary(%)")
df1=df1.drop(df.index[0:4]).reset_index(drop = True)
df1.head(5)
```

```python
#Employment with less than a Tertiary education:
df2=pd.read_excel("C:\\Users\\ibsi\\Desktop\\STEM-HTEC-Emplyoment-in-the-EU\\Datasets\\Employment in HTEC.xlsx", sheet_name="<Tertiary(%)")
df2=df2.drop(df.index[0:4]).reset_index(drop = True)
df2.head(5)
```

```python
#Employment with a Tertiary education:
df3=pd.read_excel("C:\\Users\\ibsi\\Desktop\\STEM-HTEC-Emplyoment-in-the-EU\\Datasets\\Employment in HTEC.xlsx", sheet_name="Tertiary(%)")
df3=df3.drop(df.index[0:4]).reset_index(drop = True)
df3.head(5)
```

```python
# Stacked bar for Bulgaria:


df1=df1.loc[df1["Country"] == "Bulgaria"]
df2=df2.loc[df2["Country"] == "Bulgaria"]
df3=df3.loc[df3["Country"] == "Bulgaria"]
#frames = [df, df1, df2,df3]
frames = [df1, df2,df3]
result = pd.concat(frames)
del result["Country"]
df = pd.DataFrame(data=result)
df=df.reset_index(drop=True)
df_transposed = df.T
df_transposed = df_transposed.rename(columns={0: "<Secondary", 1: "<Tertiary", 2: "Tertiary"})
df_transposed
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
my_colors=["LightSteelBlue", "LightGray"]
df_transposed.plot.bar(stacked=True, color =my_colors, ax=ax)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Education level (in %)', fontsize=15)
plt.title("Bulgaria", fontsize=20)
plt.savefig("C:\\Users\\ibsi\\Desktop\\Figure_8", transparent=True, bbox_inches="tight")
```
